```python
for i in EntriesGender.columns:
  print(i)
  print(EntriesGender[i].value_counts())
  print("////////////////////////////////")
```

```python
## checking if sum of male and female is equal to total or any error in data
count=0
for i in range(0,46):
  if EntriesGender.Male[i]+EntriesGender.Female[i]!=EntriesGender.Total[i]:
    count+=1
print(count)
```

```python
# Sports with maximum number of Entries
EntriesGender_sort=EntriesGender.sort_values(by='Total', ascending=False)
max_entries=EntriesGender_sort.iloc[:,[0,3]].head(10)
max_entries.plot(kind='barh',color="green")
plt.title("Sports with maximum number of Entries",color="Yellow")
plt.xlabel("Discipline")
plt.ylabel("Total")
plt.show()
```

```python
# Sports with top 10 sports with female entries
EntriesGender_sort=EntriesGender.sort_values(by='Female', ascending=False)
top_female=EntriesGender_sort.iloc[:,[0,1]].head(10)


```

```python
# Sports with top 10 sports with  male entries
EntriesGender_sort=EntriesGender.sort_values(by='Male', ascending=False)
top_male=EntriesGender_sort.iloc[:,[0,2]].head(10)
```

```python
## Sports where male Entries are more then Female Entries
Higher_Male_Entries=EntriesGender[EntriesGender["Female"]<EntriesGender["Male"]].sort_values(by="Male",ascending=False).iloc[:,[0,1,2]]


## double bar
bar_width = 0.35
index = range(len(Higher_Male_Entries))

plt.bar(index, Higher_Male_Entries['Female'], width=bar_width, label='Female', color='skyblue')
plt.bar([i + bar_width for i in index], Higher_Male_Entries['Male'], width=bar_width, label='Male', color='orange')

plt.xlabel('Discipline')
plt.ylabel('Number of Entries')
plt.title('Number of Entries by Discipline and Gender')
plt.xticks([i + bar_width / 2 for i in index], Higher_Male_Entries['Discipline'], rotation=45, ha='right')
plt.legend()

plt.show()
```

```python
## Sports where Female Entries are more then Female Entries
Higher_Female_Entries=EntriesGender[EntriesGender["Female"]>EntriesGender["Male"]].sort_values(by="Female",ascending=False).iloc[:,[0,1,2]]



bar_width = 0.35
index = range(len(Higher_Female_Entries))

plt.bar(index, Higher_Female_Entries['Female'], width=bar_width, label='Female', color='skyblue')
plt.bar([i + bar_width for i in index], Higher_Female_Entries['Male'], width=bar_width, label='Male', color='green')

plt.xlabel('Discipline')
plt.ylabel('Number of Entries')
plt.title('Number of Entries by Discipline and Gender')
plt.xticks([i + bar_width / 2 for i in index], Higher_Female_Entries['Discipline'], rotation=45, ha='right')
plt.legend()

plt.show()
```
